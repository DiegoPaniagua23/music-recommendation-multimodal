```python
import pandas as pd
import json
```

```python
df_final_dataset = pd.read_csv('../data/spotify-kaggle/interim/lastfm_spotify_merged.csv')
```

```python
unique_tracks = df_final_dataset['track_id'].unique()
```

```python
item_id_mapper = {track_id: i + 1 for i, track_id in enumerate(unique_tracks)}
```

```python
id_to_item_mapper = {i: track_id for track_id, i in item_id_mapper.items()}
```

```python
with open('../data/spotify-kaggle/interim/item_id_mapper.json', 'w') as f:
    json.dump(item_id_mapper, f)
```

```python
with open('../data/spotify-kaggle/interim/id_item_mapper.json', 'w') as f:
    json.dump(id_to_item_mapper, f)
```
